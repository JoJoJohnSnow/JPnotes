```python
import pandas as pd
import numpy as np
```

# DatetimeIndex主要用途是作为pandas对象的索引

# 1. DatetimeIndex 索引方法

# 1.1切片索引

可以用str类型标记或datetime-like的类型进行标记

```python
start = '2011-1-1'
end = '2011-12-31'
rng = pd.date_range(start, end, freq='10D') 
rng
```

```python
ts = pd.Series(np.random.randn(len(rng)), index=rng)
print(ts,'\n\n',ts.index)

print(ts[:5])   #切前5个
print(ts[[1,2,6]].index) #切第2、3、7个时间戳数据；！！！但是切出的数据会丢失频率参数
print(ts[::2])  #间隔一位切片
print(ts[::-1]) #反转序列
print(ts['2011-2-3':'2011-9-23'])     # 一段时间(容忍非索引精确)
print(ts['2011-5-1':]) #某段时间后
print(ts['2011-2']) #切取2月份的数据
```

```python
# pd.dataframe也同理，操作一样
ts = pd.DataFrame(np.random.randn(len(rng)), index=rng)
print(ts,'\n\n',ts.index)

print(ts[:5].index)   #切前5个
print(ts[::2].index)  #间隔一位切片
print(ts[::-1].index) #反转序列
print(ts['2011-2-3':'2011-9-23'].index)     # 一段时间(容忍非索引精确)
print(ts['2011-5-1':].index) #某段时间后
print(ts['2011-2'].index) #切取2月份的数据(不同分辨率)index构建时分辨率有两种模式（最细到day|最细到秒），不同模式下分辨率不能取比它更小的尺度）
```
